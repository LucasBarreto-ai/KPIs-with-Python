**KPIs**

**1. Importando biblioteca(s) necessária(s)**

```python
import pandas as pd
```

**2. Criar ou trazer um dataset**

```python
df = pd.DataFrame({
     "cliente_id": [1, 2, 3, 1, 2],
    "data": pd.to_datetime(["2024-01-01", "2024-01-01", "2024-01-02", "2024-01-10", "2024-01-15"]),
    "receita": [100, 200, 150, 120, 180],
    "custo": [60, 120, 90, 70, 110]
})
```

**3. Verificação de informações iniciais**

```python
df.head() # Iniciais
df.info() # Verificando se os tipos das colunas estão corretos
df.describe() # Valores estatísticos iniciais
df.isnull().sum() # Não há valores ausentes

df['data'] = pd.to_datetime(df['data'])
```

**4. KPIS essenciais**

```python
# Receita total
receita_total = df['receita'].sum()
```

```python
# Lucro total
df['lucro'] = df['receita']-df['custo']
lucro_total = df['lucro'].sum()
```

```python
# Margem de lucro
margem_lucro = (lucro_total / receita_total) * 100
```

```python
# Número de clientes únicos
df['cliente_id'].nunique()
```

```python
# Frequência de compras por cliente
df['cliente_id'].value_counts()
```

```python
# Ticket médio
ticket_medio = receita_total/len(df)
```

```python
# Frequência média de compras por clientes
frequencia_cliente = len(df)/df['cliente_id'].nunique()
frequencia_cliente
```

```python
# Receita por Dia
receita_dia = df.groupby('data')['receita'].sum()
receita_dia
```

```python
# CAC 
marketing_custo = 550
cac_cliente = marketing_custo/df['cliente_id'].nunique()
cac_cliente
```
